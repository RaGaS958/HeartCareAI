```python
cat_features=[ 'is_Male', 'FastingBS','ChestPainType_ASY',
         'ChestPainType_ATA', 'ChestPainType_NAP',
       'ChestPainType_TA', 'RestingECG_LVH', 'RestingECG_Normal', 'RestingECG_ST','ExerciseAngina_N',
       'ExerciseAngina_Y',  'ST_Slope_Down','ST_Slope_Flat', 'ST_Slope_Up']
```

```python
from scipy.stats import chi2_contingency

alpha = 0.05

df_clean['HeartDisease_bin'] = df_clean['HeartDisease']
chi2_results = {}

for col in cat_features:
    contingency = pd.crosstab(df_clean[col], df_clean['HeartDisease_bin'])
    chi2_stat, p_val, _, _ = chi2_contingency(contingency)
    decision = 'Reject Null (Keep Feature)' if p_val < alpha else 'Accept Null (Drop Feature)'
    chi2_results[col] = {
        'chi2_statistic': chi2_stat,
        'p_value': p_val,
        'Decision': decision
    }

chi2_df = pd.DataFrame(chi2_results).T
chi2_df = chi2_df.sort_values(by='p_value')
chi2_df
```

```python
for i in cat_features:
    plt.figure(figsize=(7,7))
    sns.countplot(x=df_clean[i],hue=df_clean['HeartDisease'])
```
